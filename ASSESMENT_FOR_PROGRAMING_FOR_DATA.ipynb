```python
def missing_values_table(df1):
    # Total missing values
    mis_val = df1.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * mis_val / len(df1)

    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)

    mis_val_table = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})

    # Sort the table by percentage of missing descending
    mis_val_table = mis_val_table.sort_values('% of Total Values', ascending=False)

    return mis_val_table

missing_values = missing_values_table(df1)
display(missing_values.style.background_gradient(cmap='Blues'))
```

```python
# Count duplicate rows
duplicates = df1.duplicated().sum()
print("Duplicate rows:", duplicates)

# Remove duplicates if any
if duplicates > 0:
    df1 = df1.drop_duplicates()
    print("Duplicates removed! New shape:", df1.shape)
```

```python
# formating the data time column
df1['Date'] = pd.to_datetime(df1['Date'], format='%d/%m/%Y', errors='coerce')
```

Now, let's handle the missing values. For numerical columns, I'll impute with the median. For the categorical `AQI_Bucket` column, I'll use the mode.

```python
for col in ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene', 'AQI']:
    if df1[col].isnull().any():
        median_val = df1[col].median()
        df1[col] = df1[col].fillna(median_val)

# Fill missing values in 'AQI_Bucket' with the mode
if df1['AQI_Bucket'].isnull().any():
    mode_val = df1['AQI_Bucket'].mode()[0]
    df1['AQI_Bucket'] = df1['AQI_Bucket'].fillna(mode_val)

display(missing_values_table(df1).style.background_gradient(cmap='Blues'))
```

```python
pollutants_list = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
city_pollutant_sum = df1.groupby('City')[pollutants_list].sum()

# Calculate the total sum of all pollutants for each city
city_pollutant_sum['Total_Pollutants'] = city_pollutant_sum.sum(axis=1)

# Sort to find the city with the highest total pollutants
highest_pollutant_city = city_pollutant_sum.sort_values(by='Total_Pollutants', ascending=False)

print("Cities by total pollutant levels (highest to lowest):\n", highest_pollutant_city[['Total_Pollutants']].head())

# Get the city with the absolute highest total pollutant level
most_polluted_city = highest_pollutant_city.index[0]
max_total_pollutants = highest_pollutant_city.iloc[0]['Total_Pollutants']

print(f"\nThe city with the highest total pollutant level is: {most_polluted_city} with a total of {max_total_pollutants:.2f} units.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize total pollutant levels per city
plt.figure(figsize=(15, 8))
sns.barplot(x=highest_pollutant_city.index, y=highest_pollutant_city['Total_Pollutants'], palette='viridis')
plt.title('Total Pollutant Levels Across Cities (Highest to Lowest)')
plt.xlabel('City')
plt.ylabel('Total Pollutants (Sum of all pollutants)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
